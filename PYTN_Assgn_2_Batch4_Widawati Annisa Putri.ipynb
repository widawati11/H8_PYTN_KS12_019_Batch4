```python
# Pengisian data NaN pada kolom bertipe float dengan mean
def imputasi(data_input):
    col = data_input.select_dtypes(float).columns
    class_col = data_input.columns[0]

    # Melakukan pengisian data kosong dengan mean
    for i in col:
        data_input[i].fillna(data_input.groupby(class_col)[i].transform("mean").round(1), inplace=False)

    return data_input
```

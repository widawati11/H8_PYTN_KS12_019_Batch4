```python
#Membuat mini map
mini_map = folium.Map(location=(51.507351, -0.127758), zoom_start=10)

#plugin untuk mini map
minimap=plugins.MiniMap(toggle_display=True)                                     

#memasukkan mini map ke dalam map
mini_map.add_child(minimap)

#Menambahkan fitur zoom toggler pada map
plugins.ScrollZoomToggler().add_to(mini_map)

#Menambahkan fitur full screen pada map
plugins.Fullscreen(position='topright').add_to(mini_map)                                
                                
mini_map                                
```

```python
#Peta wilayah dengan kasus kriminal terendah dan tertinggi di London
Borough = {'City of London': [51.513389, -0.088981],
           'Lambeth':[51.495315, -0.116239]}

#Membuat map
MapLondon = folium.Map(location=[51.507351, -0.127758], zoom_start=12)

#Plot lokasi
for i in Borough.items():
    folium.Marker(location=i[1], popup=i[0]).add_to(MapLondon)
    print(i)

MapLondon
```
